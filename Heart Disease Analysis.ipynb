#### Try to create Bins of Younger, Middle, Elder.

```python
data.age.min()
```

```python
data.age.max()
```

```python
#### Younger = 29 to 40
#### Middle = 40 to 50
#### Elder = 50 to 60 
#### Older = >60
```

```python
younger = data[(data['age']>= 29) & (data['age'] <= 40)]
middle = data[(data['age']> 40) & (data['age'] <=55)]
elder = data[(data['age']>55) ]
```

```python
#### 4 bins data count
print("Younger Age",len(younger))
print("Middle Age",len(middle))
print("Elder Age",len(elder))
```

```python
#Bar Plot Creation of Age Category using MatplotLib and Seaborn
plt.figure(figsize=(10,6))

sns.barplot(x=['Young_Ages','Middle_Ages','Elderly_Ages'], y=[len(younger), len(middle), len(elder)], palette='YlGn_r')
plt.title('Age Distribution', fontsize=17)
plt.xlabel('Age Range', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()
```

#### Check Male, Female data in dataset

```python
def setCategoricalValueToSex(value):
    if value ==1:
        return "Male"
    else : return "Female"
```

```python
data['SexCategorical'] = data['sex'].apply(setCategoricalValueToSex)
```

```python
def heartDiseaseCategorical(value):
    if value ==1:
        return "Presence"
    else : return "Absence"
data['HeartDisease'] = data['num'].apply(heartDiseaseCategorical)
```

```python
data.head(3)
```

```python
def setAgeRanges(value):
    if value >= 29 and value <= 40 :
        return "Younger"
    elif value >= 40 and value <= 55:
        return "Middle"
    else : return "Elder"
```

```python
data['AgeCategorical'] = data['age'].apply(setAgeRanges)
```

```python
data.head(3)
```

```python
#Swarm Plot Creation of Gender Based Age Category using MatplotLib and Seaborn

plt.figure(figsize=(11,8))
sns.swarmplot(x='AgeCategorical', y='age', hue='SexCategorical', data=data, order=['Younger','Middle','Elder'], palette='Oranges_r')
plt.title('Gender Based Age Category', fontsize=17)
plt.xlabel('Age Category', fontsize=15)
plt.ylabel('Age', fontsize=15)
plt.show()
```
